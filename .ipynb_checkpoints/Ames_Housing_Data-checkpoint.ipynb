# Ames Housing Data

Per https://github.com/melindaleung/Ames-Iowa-Housing-Dataset:

> The Ames Housing dataset, which describes the sale of individual residential property in Ames, Iowa from 2006 to 2010, was compiled by Dean De Cock for use in data science education. It was designed based after the Boston Housing dataset and is now considered a more modernized and expanded version of it. More details of this dataset are described in [Ames, Iowa: Alternative to the Boston Housing Data as an End of Semester Regression Project](https://ww2.amstat.org/publications/jse/v19n3/decock.pdf).

The GitHub project includes a [description of the dataset](https://raw.githubusercontent.com/melindaleung/Ames-Iowa-Housing-Dataset/master/data/data%20description.txt).


````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import inflection

random_state = 42

def camel_to_snake(s):
    s = inflection.underscore(s)
    s = ''.join(['_' + c if c.isdigit() and n else c for n, c in enumerate(s)])
    return s

df = pd.read_csv("https://raw.githubusercontent.com/melindaleung/Ames-Iowa-Housing-Dataset/master/data/ames%20iowa%20housing.csv", index_col="Id")
#df = df.fillna('NA')

df.columns = [camel_to_snake(c) for c in df.columns.tolist()]

df["central_air"] = df["central_air"] == "Y"
df["paved_drive"] = df["paved_drive"] == "Y"

na_columns = ["alley", "bsmt_qual","bsmt_cond","bsmt_exposure","bsmt_fin_type_1","bsmt_fin_type_2","fireplace_qu","garage_type","garage_finish","garage_qual","garage_cond","pool_qc","fence","misc_feature"]
df[na_columns] = df[na_columns].fillna('NA')

df = df.dropna()
df[["lot_frontage","garage_yr_blt"]] = df[["lot_frontage","garage_yr_blt"]].astype(int)

df_full_train, df_test = train_test_split(df, test_size=.2, random_state=random_state,)
df_train, df_val = train_test_split(df_full_train, test_size=.25, random_state=random_state)

df_full_train = df_full_train.reset_index(drop=True)
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

````

````python
corr = df_full_train[numerical_columns].corrwith(np.log1p(df_full_train.sale_price)).abs()
pd.DataFrame(corr[(corr >= .1)].sort_values(ascending=False))
````

````python
y_train = np.log1p(df_train["sale_price"])
y_val = np.log1p(df_val["sale_price"])
y_test = np.log1p(df_test["sale_price"])
y_full_train = np.log1p(df_full_train["sale_price"])

columns_to_drop = ["sale_price"]

categorical_columns = [column for column in categorical_columns if column not in columns_to_drop]
df_full_train = df_full_train.drop(columns_to_drop, axis=1)
df_train = df_train.drop(columns_to_drop, axis=1)
df_val = df_val.drop(columns_to_drop, axis=1)
df_test = df_test.drop(columns_to_drop, axis=1)
````

````python
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=False)
dv.fit(df[categorical_columns + numerical_columns].to_dict(orient='records'))

full_train_dict = df_full_train[categorical_columns + numerical_columns].to_dict(orient='records')
X_full_train = dv.transform(full_train_dict)

train_dict = df_train[categorical_columns + numerical_columns].to_dict(orient='records')
X_train = dv.transform(train_dict)

val_dict = df_val[categorical_columns + numerical_columns].to_dict(orient='records')
X_val = dv.transform(val_dict)

test_dict = df_test[categorical_columns + numerical_columns].to_dict(orient='records')
X_test = dv.transform(test_dict)
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_val)

score = model.score(X_val, y_val)
print(score)

rmse = mean_squared_error(y_val, y_pred, squared=False)
rmse
````

````python
from sklearn.linear_model import Ridge

for a in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=a,random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    
    score = np.sqrt(mean_squared_error(y_val, y_pred))
    
    print(f"alpha: {a}, rmse:{round(score, 3)}")
````

````python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_text

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)

print(export_text(dt, feature_names=dv.get_feature_names()))
````

````python
y_pred = dt.predict(X_val)
mean_squared_error(y_pred, y_val, squared=False)
````

````python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=10, n_jobs=-1, random_state=random_state)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)

mean_squared_error(y_val, y_pred, squared=False)
````

````python
results = []

rf = RandomForestRegressor(n_jobs=-1, random_state=random_state, warm_start=True)

for n_estimators in range(10, 201, 10):
  rf.n_estimators = n_estimators
  rf.fit(X_train, y_train)
  y_pred = rf.predict(X_val)

  results.append((n_estimators, mean_squared_error(y_val, y_pred, squared=False)))

results
````

````python
import matplotlib.pyplot as plt

df_results = pd.DataFrame(results, columns=['n_estimators', 'rmse'])
plt.plot(df_results.n_estimators, df_results.rmse)
````

````python
import xgboost as xgb

features = dv.get_feature_names()
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)

watchlist = [(dtrain, 'train'), (dval, 'val')]

xgb_params = {
    'eta': 0.3, 
    'max_depth': 6,
    'min_child_weight': 1,

    'objective': 'reg:squarederror',
    'nthread': 8,

    'seed': 1,
    'verbosity': 1,
}

xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist)
#y_pred = model.predict(dval)
````

````python

evals = [(dval, 'val')]

scores = []
for eta in [0.3, 0.1, 0.01]:
  xgb_params["eta"] = eta
  bst = xgb.train(xgb_params, dtrain, num_boost_round=500, evals=evals, early_stopping_rounds=10)
  scores.append((eta, bst.best_score, bst.best_iteration))
````

````python
df_scores = pd.DataFrame(scores, columns=["eta", "best_score", "best_iteration"])
df_scores.sort_values("best_score")
````

````python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso, ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('scale', StandardScaler()),
    ('regr', LinearRegression())
])

param_grid = [
    {
        'regr': [Lasso(max_iter=4000)],
        'regr__alpha': np.logspace(-4, 1, 6),
        #'regr__l1_ratio' : [0,0.25,0.5,0.75,1]
    },
        {
        'regr': [ElasticNet(max_iter=4000)],
        'regr__alpha': np.logspace(-4, 1, 6),
        'regr__l1_ratio' : [0,0.25,0.5,0.75,1]
    },
]

grid = GridSearchCV(pipe, param_grid=param_grid, cv=10, n_jobs=-1, verbose=2)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_estimator_)
print(grid.best_index_)
print(grid.best_score_)

print('Score:\t{}'.format(grid.score(X_val, y_val)))

y_pred = grid.predict(X_val)
rmse = mean_squared_error(y_val, y_pred, squared=False)
rmse
````

In XGBoost, the colsample_bytree parameter controls the number of features we select for each tree â€” itâ€™s similar to max_features from random forest. Experiment with this parameter and see if it improves the performance: try values from 0.1 to 1.0 with step 0.1. Usually the optimal values are between 0.6 and 0.8, but sometimes 1.0 gives the best result.

In addition to randomly selecting columns (features), we can also select a subset of rows (customers). This is called subsampling and it helps to prevent overfitting. In XGBoost, the subsample parameter controls the fraction of examples we select for training each tree in the ensemble. Try values from 0.4 to 1.0 with step 0.1. Usually the optimal values are between 0.6 and 0.8.

````python
colsample_by_tree = np.linspace(0.0, 1.0, 11) 
subsample = np.arange(0.4, 1.1, .1)

colsample_by_tree, subsample
````
